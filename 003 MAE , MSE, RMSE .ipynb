```python
from sklearn.metrics import mean_squared_error
```

```python
y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]
mse = mean_squared_error(y_true, y_pred)
print("MSE: %f\n" % mse)
```

```python
y_true = [[0.5, 1],[-1, 1],[7, -6]]
y_pred = [[0, 2],[-1, 2],[8, -5]]
mse = mean_squared_error(y_true, y_pred)
print("MSE: %f\n" % mse)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]
mae = mean_absolute_error(y_true, y_pred)
print("MAE: %f\n" % mae)

```

```python

y_true = [[0.5, 1], [-1, 1], [7, -6]]
y_pred = [[0, 2], [-1, 2], [8, -5]]
mae = mean_absolute_error(y_true, y_pred)
print("MAE: %f\n" % mae)
```
